```python
from inference import StableDiffusion
```

```python
prompt = "A tree house lin the woods, with a swing hanging from a tree, art"
model = StableDiffusion()
latent = model.generate(prompt, seed=221, steps=50)
```

```python
image = model.decode(latent)
model.show(image)[0]
```

```python
print("latent shape", latent.shape)
print("encoding shape", model.encode_text([prompt]).shape)
print("embedding shape", model.get_text_embedding(prompt, batch_size=1).shape)
```
